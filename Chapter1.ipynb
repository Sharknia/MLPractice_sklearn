```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("iris의 키 : \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:193] + "\n....")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print("X_train 크기 : {}".format(X_train.shape))
```

```python
import pandas as pd
import sys
import matplotlib
```

```python
import numpy as np
```

```python
import scipy as sp
import IPython
import sklearn
import mglearn
```

```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(14, 14), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape : {}".format(X_new.shape))
```

```python
prediction = knn.predict(X_new)
print("예측 : {}".format(prediction))
print("예측한 타깃의 이름 : {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_pred = knn.predict(X_test)
print("테스트 세트에 대한 예측값 : \n{}".format(y_pred))
```

```python
print("테스트 세트의 정확도:{:.2f}".format(np.mean(y_pred==y_test)))
```

```python
print("테스트 세트의 정확도 : {:.2f}".format(knn.score(X_test, y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)   #데이터셋에서 75프로는 훈련데이터, 25프로는 테스트 데이터로 구분
knn = KNeighborsClassifier(n_neighbors=1)   #k-최근접 이웃 알고리즘
knn.fit(X_train, y_train)   #knn 객체 자체를 반환하고, 객체 자체를 변경시킴
print("테스트 세트 정확도 : {:.2f}\n".format(knn.score(X_test, y_test)))    #테스트 데이터를 넣어서 정확도 확인 가능
```
